```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
data = [
    [0.5, 0.5],
    [0.7, 0.3],
    [0.9, 0.6]
]
labels = [0, 1, 1]

data_arr = np.array(data)
data_arr
```

```python
# plot the data
sns.scatterplot(x=data_arr[:,0], y=data_arr[:,1], hue=labels)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter plot of the data")
plt.show()
```

```python
initial_value = -100
coeffs = np.array([initial_value, initial_value])
bias = 0.0  # Fix bias to visualize a 2D error surface over (a1, a2)

def composite_function(x, coeffs, bias):
    if len(x) != len(coeffs):
        raise ValueError("Input length must match the number of coefficients")
    z = np.sin(np.dot(x, coeffs) + bias)
    return z

# First sample
x0 = data_arr[0]
y0 = labels[0]

# Parameter grid for the two coefficients
possible_values = np.linspace(-10, 10, 250)
A1, A2 = np.meshgrid(possible_values, possible_values)

# Model output and squared error surface
Z = np.sin(x0[0] * A1 + x0[1] * A2 + bias)
error_surface = (Z - y0) ** 2

# Contour plot
plt.figure(figsize=(8, 6))
contour = plt.contourf(A1, A2, error_surface, levels=40, cmap="viridis")
plt.colorbar(contour, label="Squared Error")
plt.xlabel("Coefficient a1")
plt.ylabel("Coefficient a2")
plt.title("Error Surface (First Sample, Fixed Bias = 0)")
plt.show()

# 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")
surf = ax.plot_surface(A1, A2, error_surface, cmap="viridis", linewidth=0, antialiased=True)
fig.colorbar(surf, shrink=0.6, aspect=12, label="Squared Error")
ax.set_xlabel("Coefficient a1")
ax.set_ylabel("Coefficient a2")
ax.set_zlabel("Squared Error")
ax.set_title("3D Error Surface (First Sample, Fixed Bias = 0)")
plt.show()
```
